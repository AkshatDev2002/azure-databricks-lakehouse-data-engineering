# Dynamic Capabalities

```python
dbutils.widgets.text("file_name", "")
```

```python
p_file_name = dbutils.widgets.get("file_name")
```

# **Data Consumption**

```python
df = spark.readStream.format("cloudFiles")\
.option("cloudFiles.format", "parquet")\
.option("cloudFiles.schemaLocation", f"abfss://bronze@dbxdev.dfs.core.windows.net/checkpoint_{p_file_name}")\
.load(f"abfss://dbsource@dbxdev.dfs.core.windows.net/{p_file_name}")
```

# **Data Persistence**

```python
df.writeStream.format("parquet")\
    .outputMode("append")\
    .option("checkpointLocation",f"abfss://bronze@dbxdev.dfs.core.windows.net/checkpoint_{p_file_name}")\
    .option("path",f"abfss://bronze@dbxdev.dfs.core.windows.net/{p_file_name}")\
    .trigger(once=True)\
    .start()
```

```python
df = spark.read.format("parquet").load(f"abfss://bronze@dbxdev.dfs.core.windows.net/{p_file_name}")

display(df)
```
