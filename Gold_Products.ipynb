# **DLT Pipeline**

```python
import dlt
from pyspark.sql.functions import *
```

## **Streaming Table**

```python
#Expectations

my_rules = {
    "rule1": "proudct_id is not null",
    "rule2": "product_name is not null"
}
```

```python
@dlt.table()

@dlt.expect_all_or_drop(my_rules)
def DimProducts_stage():
  df =  spark.readStream.table("azuredb_catalog.silver.products_silver")

  return df
```

## **Streaming View**

```python
@dlt.view

def DimProduct_view():
  df = spark.readStream.table("Live.DimProducts_view")
  return df
```

**DimProducts**

```python
dlt.create_streaming_table("DimProduct")
```

```python
dlt.apply_changes(
  target = "DimProduct",
  source = "Live.DimProducts_view",
  keys = ["product_id"],
  sequence_by = "product_id", 
  stored_as_scd_type = 2 )
```
