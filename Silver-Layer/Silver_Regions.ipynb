# **Data Consumption**

```python
df = spark.read.table("azuredb_catalog.bronze.regions")\
    
df.display()
```

```python
df = df.drop("_rescued_data")
```


# **Data Persistence**


```python
df.write.format("delta")\
    .mode("overwrite")\
    .save("abfss://silver@dbxdev.dfs.core.windows.net/regions")
```
