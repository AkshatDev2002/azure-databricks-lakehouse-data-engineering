# **Fact Orders**

### Data Reading

```python
df = spark.sql("select * from azuredb_catalog.silver.orders_silver")
df.display()
```

```python
df_dimcus = spark.sql("select DimCustomerKey, customer_id as dim_customer_id from azuredb_catalog.gold.dimcustomers")

df_dimpro = spark.sql("select product_id as DimProductKey, product_id as dim_product_id from azuredb_catalog.gold.dimproducts")
```

**Fact DataFrame**

```python
df_fact = df.join(df_dimcus, df['customer_id'] == df_dimcus['dim_customer_id'], how='left').join(df_dimpro, df['product_id'] == df_dimpro['dim_product_id'], how='left')

df_fact_new = df_fact.drop('dim_customer_id','dim_product_id','customer_id','product_id')
```

```python
df_fact_new.display()
```

# **Upsert on Fact Table**

```python
from delta.tables import DeltaTable
```

```python
if spark.catalog.tableExists("azuredb_catalog.gold.FactOrders"):

  dlt_obj = DeltaTable.forName(spark, "azuredb_catalog.gold.FactOrders")

  dlt_obj.alias("trg").merge(df_fact_new.alias("src"), "trg.order_id = src.order_id AND trg.DimCustomerKey = src.DimCustomerKey AND trg.DimProductKey = src.DimProductKey")\
    .whenMatchedUpdateAll()\
    .whenNotMatchedInsertAll()\
    .execute()
else:
  df_fact_new.write.format("delta")\
      .option("path","abfss://gold@dbxdev.dfs.core.windows.net/FactOrders")\
      .saveAsTable("azuredb_catalog.gold.FactOrders")
```
