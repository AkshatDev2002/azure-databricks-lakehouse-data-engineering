### Data Reading

```python
df = spark.read.format("parquet").load("abfss://dbsource@dbxdev.dfs.core.windows.net/orders")
df.display()
```

```python
df = spark.readStream.format("cloudFiles")\
.option("cloudFiles.format", "parquet")\
.option("cloudFiles.schemaLocation", "abfss://bronze@dbxdev.dfs.core.windows.net/checkpoint_orders")\
.load("abfss://dbsource@dbxdev.dfs.core.windows.net/orders")
```

```python
df.writeStream.format("parquet")\
    .outputMode("append")\
    .option("checkpointLocation", "abfss://bronze@dbxdev.dfs.core.windows.net/checkpoint_orders")\
    .option("path", "abfss://bronze@dbxdev.dfs.core.windows.net/orders")\
    .trigger(once=True)\
    .start()
```
